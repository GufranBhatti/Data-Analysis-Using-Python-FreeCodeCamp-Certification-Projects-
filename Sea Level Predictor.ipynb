```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.tail()
```

```python
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],edgecolors='g')
```

```python
result = linregress(df['Year'],df['CSIRO Adjusted Sea Level'])
predict = np.arange(df['Year'].min(),2050,1)
line = predict*result.slope + result.intercept
plt.plot(predict,line)
```

```python
new_df = df[df['Year']>=2000]
new_result = linregress(new_df['Year'], new_df['CSIRO Adjusted Sea Level'])
new_predict = np.arange(new_df['Year'].min(), 2050, 1)
new_line = new_predict * new_result.slope + new_result.intercept
plt.plot(new_predict, new_line)
```

```python
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
```
